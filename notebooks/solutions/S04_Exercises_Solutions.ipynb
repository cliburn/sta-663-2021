**Exercise 1**.

- Create the `mtcars` DataFrame from this URL `https://bit.ly/3oAZyrd`
- Find the mean `wt` by number of cylinders sorted in decreasing order, keeping the result as a DataFrame.
- Find all car models that have numbers in the name
- Create a correlation matrix of `hp`, `weight` and `qsec` for cars with 4 gears

```python
import pandas as pd
```

```python
url = 'https://bit.ly/3oAZyrd'

mtcars = pd.read_csv(url)
```

```python
mtcars.head(3)
```

```python
mtcars.groupby('cyl')[['wt']].mean().sort_values('wt', ascending=False)
```

```python
mtcars.model[mtcars.model.str.contains(r'\d+')]
```

```python
mtcars[mtcars.gear==4][['hp', 'wt', 'qsec']].corr()
```

**Exercise 2**.

- Convert the data in `mess.txt` to the DataFrame in `tidy.csv` using Pandas operations


```python
df = pd.read_csv('mess.txt', sep=' ')
```

```python
final = pd.read_csv('tidy.csv')
```

```python
df[['name', 'age', 'site']] = df.desc.str.split('-', expand=True)
```

```python
df = df.drop(columns='desc')
```

```python
df = pd.melt(df, id_vars=['name', 'age', 'site'], var_name='gene', value_name='expression')
```

```python
df['gene'] = df.gene.str.replace('gene', '')
```

```python
df.sort_values('name')
```

**Exercise 3**.

- Repeat exercise 2 as a single chained command. (Challenging)

```python
(
    pd.read_csv('mess.txt', sep=' ').
    assign(
        name = lambda x: x.desc.str.split('-').str[0],
        age = lambda x: x.desc.str.split('-').str[1],
        site = lambda x: x.desc.str.split('-').str[2]
    ).drop(columns=['desc']).
    melt(id_vars=['name', 'age', 'site'], var_name='gene', value_name='expression').
    assign(gene=lambda x: x.gene.str.replace('gene', '')).
    sort_values('name')
)
```
